```python
# importing libraries
import pandas as pd
```

```python
wh_2015 = pd.read_csv("../data/raw/2015.csv")
wh_2016 = pd.read_csv("../data/raw/2016.csv")
wh_2017 = pd.read_csv("../data/raw/2017.csv")
wh_2018 = pd.read_csv("../data/raw/2018.csv")
wh_2019 = pd.read_csv("../data/raw/2019.csv")
```

```python
# renaming columns
wh_2019.rename(columns={
    "Overall rank": "rank",
    "Country or region": "country",
    "Score": "score",
    "GDP per capita": "gdp_per_capita",
    "Social support": "social_support",
    "Healthy life expectancy": "life_expectancy",
    "Freedom to make life choices": "freedom",
    "Generosity": "generosity",
    "Perceptions of corruption": "corruption"
}, inplace=True)

wh_2018.rename(columns={
    "Overall rank": "rank",
    "Country or region": "country",
    "Score": "score",
    "GDP per capita": "gdp_per_capita",
    "Social support": "social_support",
    "Healthy life expectancy": "life_expectancy",
    "Freedom to make life choices": "freedom",
    "Generosity": "generosity",
    "Perceptions of corruption": "corruption"
}, inplace=True)

wh_2017.rename(columns={
    "Happiness.Rank": "rank",
    "Country": "country",
    "Happiness.Score": "score",
    "Economy..GDP.per.Capita.": "gdp_per_capita",
    "Family": "social_support",
    "Health..Life.Expectancy.": "life_expectancy",
    "Freedom": "freedom",
    "Generosity": "generosity",
    "Trust..Government.Corruption.": "corruption"
}, inplace=True)

wh_2016.rename(columns={
    "Happiness Rank": "rank",
    "Country": "country",
    "Happiness Score": "score",
    "Economy (GDP per Capita)": "gdp_per_capita",
    "Family": "social_support",
    "Health (Life Expectancy)": "life_expectancy",
    "Freedom": "freedom",
    "Generosity": "generosity",
    "Trust (Government Corruption)": "corruption"
}, inplace=True)

wh_2015.rename(columns={
    "Happiness Rank": "rank",
    "Country": "country",
    "Happiness Score": "score",
    "Economy (GDP per Capita)": "gdp_per_capita",
    "Family": "social_support",
    "Health (Life Expectancy)": "life_expectancy",
    "Freedom": "freedom",
    "Generosity": "generosity",
    "Trust (Government Corruption)": "corruption"
}, inplace=True)
```

```python
# adding a new year column
wh_2019["year"] = 2019
wh_2018["year"] = 2018
wh_2017["year"] = 2017
wh_2016["year"] = 2016
wh_2015["year"] = 2015
```

```python
# subsetting the dataframes to join
wh_2019 = wh_2019[["year", "rank", "country", "score", "gdp_per_capita", "social_support", "life_expectancy", "freedom", "generosity", "corruption"]]
wh_2018 = wh_2018[["year", "rank", "country", "score", "gdp_per_capita", "social_support", "life_expectancy", "freedom", "generosity", "corruption"]]
wh_2017 = wh_2017[["year", "rank", "country", "score", "gdp_per_capita", "social_support", "life_expectancy", "freedom", "generosity", "corruption"]]
wh_2016 = wh_2016[["year", "rank", "country", "score", "gdp_per_capita", "social_support", "life_expectancy", "freedom", "generosity", "corruption"]]
wh_2015 = wh_2015[["year", "rank", "country", "score", "gdp_per_capita", "social_support", "life_expectancy", "freedom", "generosity", "corruption"]]
```

```python
# union all
df = pd.concat([wh_2019, wh_2018, wh_2017, wh_2016, wh_2015])
```
