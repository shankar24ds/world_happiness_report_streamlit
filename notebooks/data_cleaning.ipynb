```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# importing the file needed for the analysis
df = pd.read_csv("../data/processed/world_happiness_integrated.csv")
```

```python
# checking for inconsitencies in country names among years
df['country'].value_counts().reset_index().sort_values('index')
```

```python
replacements = {'Hong Kong S.A.R., China': 'Hong Kong', 'Northern Cyprus': 'North Cyprus', 'Trinidad & Tobago': 'Trinidad and Tobago', 'Taiwan Province of China': 'Taiwan'}
df['country'] = df['country'].replace(replacements)
```

```python
country_count = df['country'].value_counts().reset_index().sort_values('index')
country_list = set(country_count[country_count['country']==5]['index'])
country_list
```

```python
df = df[df['country'].isin(country_list)]
```

```python
# checking country count
df['country'].value_counts().reset_index()
```

```python
df.to_csv("../data/processed/df_cleaned.csv", index=False)
```
